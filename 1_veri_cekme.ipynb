```python
import pandas as pd
from datasets import load_dataset
```

#veri setini indirme

```python
dataset = load_dataset("maydogan/Turkish_SentimentAnalysis_TRSAv1")
```

#dataframe oluşturma

```python
df = pd.DataFrame(dataset['train'])
```

#veri setinin ilk 10 değerleri

```python
df.head(10)
```

#pozitif verilerden 15k örneklem alıyoruz ama random state yazmamız lazım yoksa her çalıştırdığımızda farklı veriler verecek.

```python
pozitif_15k = df[df['score'] == 'Positive'].sample(n=15000, random_state=42)
negatif_15k = df[df['score'] == 'Negative'].sample(n=15000, random_state=42)
notr_15k = df[df['score'] == 'Neutral'].sample(n=15000, random_state=42)

#şimdi hepsinden 15k aldıktan sonra concat ile birleştirme yapalım alt alta

df_dengeli = pd.concat([pozitif_15k, negatif_15k, notr_15k])

print("karıştırmadan önce ilk 5 satır:")
print(df_dengeli.head())

print("karıştırmadan önce son 5 satır:")
print(df_dengeli.tail())

```

#şimdi ise bu verileri karıştıralım. frac verinin % kaçının seçileceğini sağlar. reset_index ise karıştırıldıktan sonra indeksler dağınık görünür bu dağınıklığı düzenler 

```python
df_dengeli = df_dengeli.sample(frac=1, random_state=42).reset_index(drop=True)

print("karışmış hali ilk 5 satır:")
print(df_dengeli.head())

#value counts hangi veri kaç kere tekrar ediyor onu yazar.
print("\nher sınıftan kaç tane var kontrol: ")
print(df_dengeli['score'].value_counts())
```
