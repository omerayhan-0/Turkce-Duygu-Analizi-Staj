WorldCloud en popüler görselleştirme aracıdır.

#terminalden wordcloudu indiriyoruz

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_csv('temizlenmis_veri_seti.csv')
#temizlenmiş veri setini okuyoruz
```

```python
#sadece pozitifleri filtreleyelim
#boş değerleri dropna ile siliyoruz
pozitif_yorumlar = df[df['score'] == 'Positive']['temiz_review'].dropna()
```

```python
#worldcloud dataframe veya bir excel tablosu okumayı bilmez bu yüzden verileri birleştirerek bir metin haline getiriyoruz.
pozitif_metin = " ".join(pozitif_yorumlar)
```

```python
#wordcloud nesnesi oluşturuyoruz, arka plan beyaz ve harfler de pozitif olduğu için yeşil tonlarında olsun.
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    colormap='Greens'
).generate(pozitif_metin)
```

```python
#ekrana çizdirme işlemi
#figure ile grafik oluşturuyoruz
#yukarda oluşturduğumuz bir resimdi piksellerle oluşuyor, figure ile bu resme çerçeve oluşturuyoruz
#imshow ile resmi çerçeveye yerleştiriyoruz
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear') # pikselli ve çirkin görüntüyü yumuşatır.
plt.axis('off') # kenardaki x ve y eksen çizgilerini kapatıyor.

plt.title("pozitif yorumlar kelime bulutu", fontsize=18, fontweight='bold', color='green')
plt.show()
```
